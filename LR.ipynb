**Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.**

**Answer:**
Simple Linear Regression (SLR) is a supervised machine learning algorithm that models the linear relationship between one independent variable (X) and one dependent variable (Y).

**Purpose:**

(1) Prediction – Predict a continuous output (e.g., predict salary from years of experience).

(2) Understanding relationships – Quantify how much Y changes when X changes.

(3) Trend analysis – Identify the direction (positive/negative) and strength of the X-Y relationship.
The model fits the best straight line through the data: Y = b0 + b1*X, minimizing prediction errors.

**Question 2: What are the key assumptions of Simple Linear Regression?**

**Answer:**

Key Assumptions of Simple Linear Regression:
1. Linearity: The relationship between X and Y must be linear (a straight-line relationship exists).
2. Independence: Each observation/data point must be independent of others (no autocorrelation).
3. Homoscedasticity: Residuals (errors) must have constant variance for all values of X.
4. Normality of Residuals: The error terms should follow a normal distribution.
5. No significant outliers: Extreme data points can heavily distort the regression line and results.

**Question 3: Write the mathematical equation for a simple linear regression model and explain each term.**

**Answer:**

The SLR equation is: Y = b0 + b1*X + e
Where:


**Y =** Dependent variable (the output/target we want to predict)

**b0 =** Intercept (the value of Y when X = 0; where the line crosses the Y-axis)

**b1 =** Slope (the change in Y for every 1-unit increase in X; shows direction and magnitude)

**X =** Independent variable (the input/predictor used to predict Y)

**e =** Error term / Residual (the difference between actual Y and predicted Y; captures noise)

**Question 5: What is the method of least squares in linear regression?**

**Answer:**
The Method of Least Squares finds the best-fit line by minimizing the total sum of squared differences between actual values (yi) and predicted values (yi_hat).

Objective: Minimize SSE = Sum of (yi - yi_hat)^2

This gives optimal formulas:

Slope (b1) = [n*Sum(xi*yi) - Sum(xi)*Sum(yi)] / [n*Sum(xi^2) - (Sum(xi))^2]

Intercept (b0) = y_mean - b1 * x_mean

Squaring the errors ensures positive values, penalizes larger errors more, and allows calculus-based optimization to find the unique best-fit line.

**Question 6: What is Logistic Regression? How does it differ from Linear Regression?**

**Answer:**

Logistic Regression is a supervised machine learning algorithm used for classification problems — where the output (dependent variable) is categorical, typically binary (0 or 1, Yes or No, Spam or Not Spam).

Even though it has "regression" in its name, it is not used for predicting continuous values. Instead, it predicts the probability that a given input belongs to a particular class.
It uses the Sigmoid (Logistic) Function to squish any real number into a value between 0 and 1:

P(Y=1) = 1 / (1 + e^(−z)) where z = b0 + b1·X


If P ≥ 0.5 → classify as 1
If P < 0.5 → classify as 0

Differences from Linear Regression

**Nature of the Problem —** Linear Regression is used when the output variable is continuous, such as predicting salary, temperature, or house prices. Logistic Regression is used when the output is categorical, such as predicting whether a patient has a disease or not.

**Output Interpretation —** In Linear Regression, the output is a direct numerical value that can range from negative infinity to positive infinity. In Logistic Regression, the output is always a probability value between 0 and 1, which is then converted into a class label using a threshold (usually 0.5).

**Equation and Shape —** Linear Regression produces a straight line that fits the data. Logistic Regression produces an S-shaped (sigmoid) curve. This curve is essential because probabilities must be bounded, and a straight line cannot guarantee values stay within 0 and 1.

**Loss Function —** Linear Regression minimizes Mean Squared Error (MSE), which measures the average squared difference between actual and predicted values. Logistic Regression minimizes Log Loss (Binary Cross-Entropy), which penalizes wrong probability predictions heavily, making it more suitable for classification tasks.

**Decision Boundary —** Linear Regression has no concept of a decision boundary since it predicts values on a continuous scale. Logistic Regression creates a decision boundary that separates one class from another, classifying inputs as class 0 or class 1 based on which side of the boundary they fall on.



**Question 7: Name and briefly describe three common evaluation metrics for regression models.**

**Answer:**

Three Common Evaluation Metrics for Regression Models:
1. Mean Absolute Error (MAE): Average of absolute differences |actual - predicted|. Easy to interpret; treats all
errors equally. Lower is better.
2. Mean Squared Error (MSE): Average of squared differences (actual - predicted)^2. Penalizes larger errors
more heavily. Sensitive to outliers. Lower is better.
3. Root Mean Squared Error (RMSE): Square root of MSE. It is in the same unit as the target variable, making
interpretation intuitive. Lower RMSE = better model fit. Most commonly used metric.

**Question 8: What is the purpose of the R-squared metric in regression analysis?**

**Answer:**

R-squared (R²) measures the proportion of variance in the dependent variable (Y) that is explained by the
independent variable(s) in the model.
Formula: R² = 1 - (SS_residual / SS_total), where SS_residual = Sum(yi - yi_hat)^2 and SS_total = Sum(yi -
y_mean)^2.
Interpretation:
- R² = 0: The model explains none of the variance (as good as predicting the mean).
- R² = 1: The model perfectly explains all variance.
- R² = 0.85: The model explains 85% of the variance in Y.
A higher R² indicates a better fit, but it should be used alongside other metrics to avoid overfitting.

```python
'''
Question 9: Write Python code to fit a simple linear regression model using scikit-learn
'''
# Python Code to Fit SLR using scikit-learn
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
# Sample dataset
X = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
y = np.array([3, 5, 7, 9, 11, 13, 15])
# Create and fit the model
model = LinearRegression()
model.fit(X, y)
# Print slope and intercept
print("Slope (b1):", round(model.coef_[0], 4))
print("Intercept (b0):", round(model.intercept_, 4))
print("R-squared:", round(r2_score(y, model.predict(X)), 4))
```

**Question 10: How do you interpret the coefficients in a simple linear regression model?**

**Answer:**

Interpreting Coefficients in SLR model Y = b0 + b1*X:

**Intercept (b0):** The predicted value of Y when X = 0. It is the starting point (baseline value) of the regression line on the Y-axis. Example: If b0 = 20000, a person with 0 years of experience earns Rs 20,000.

**Slope (b1):** The change in Y for every 1-unit increase in X. It tells us the direction and magnitude of the relationship. Example: If b1 = 5000, each additional year of experience increases salary by Rs 5,000.

**Positive b1** = positive relationship (Y increases as X increases).

**Negative b1** = inverse relationship (Y decreases as X increases).
